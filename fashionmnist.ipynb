```python
import torch
import torchvision
from torchvision import datasets
from torchvision.transforms import ToTensor
from matplotlib import pyplot as plt
```

```python
training_data=datasets.FashionMNIST(
    root='fashionmnist_data',
    train=True,
    download=True,
    transform=ToTensor()
)
```

```python
test_data=datasets.FashionMNIST(
    root='fashionmnist_data',
    train=False,
    download=True,
    transform=ToTensor()
)
```
